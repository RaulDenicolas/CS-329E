##Inventory

```python
%%bigquery
create or replace table retails_csp.Inventory(
  Product_ID STRING not null,
  Product_Name	STRING,
  Beginning_Inventory	INTEGER,
  Inventory_Recieved	INTEGER,
  Inventory_Sold	INTEGER,
  Ending_Inventory	INTEGER,
  Predicted_Sales	INTEGER,
  data_source	STRING not null,
  load_time	TIMESTAMP not null,
  effective_time TIMESTAMP default current_timestamp() not null,
  discontinue_time TIMESTAMP,
  status_flag BOOL not null,
  primary key (Product_ID, Beginning_Inventory, Predicted_Sales, effective_time) not enforced)
  as select *, current_timestamp(), null, true
  from retails_stg.Inventory;
```

```python
%%bigquery
select c.column_name, c.data_type, c.is_nullable, c.column_default, k.constraint_name
from retails_csp.INFORMATION_SCHEMA.COLUMNS c left join retails_csp.INFORMATION_SCHEMA.KEY_COLUMN_USAGE k
on c.table_name = k.table_name and c.column_name = k.column_name
where c.table_name = 'Inventory'
order by c.ordinal_position;
```

```python
%%bigquery
select * from retails_csp.Inventory
limit 5
```
